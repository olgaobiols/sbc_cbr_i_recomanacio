```python
import json

# Carregar la base de casos
with open("base_de_casos.json", "r", encoding="utf-8") as f:
    casos = json.load(f)

# Crear un set amb tots els ingredients_en de tots els menÃºs
ingredients_en_set = set()

for cas in casos:
    sol = cas.get("solucio", {})
    for plat in ["primer_plat", "segon_plat", "postres"]:
        if plat in sol and "ingredients_en" in sol[plat]:
            ingredients_en_set.update(sol[plat]["ingredients_en"])

ingredients_en_set
```

```python
import pandas as pd

# Carregar la base de dades d'ingredients
df_ing = pd.read_csv("ingredients_en.csv")

# Convertir la columna ingredient_name en un set
ingredients_db = set(df_ing["ingredient_name"].str.strip().str.lower())

# Normalitzar ingredients del conjunt anterior (per si hi ha majÃºscules)
ingredients_to_check = {i.strip().lower() for i in ingredients_en_set}

# Comprovar quins NO estan a la base de dades
missing_ingredients = sorted(ingredients_to_check - ingredients_db)

print("Ingredients NO presents a ingredients_en.csv:")
for ing in missing_ingredients:
    print("-", ing)
```

```python
import pandas as pd

# Carregar la base d'ingredients
df_ing = pd.read_csv("ingredients_en.csv")

print("ðŸ§¾ COLUMNES I CATEGORIES DISPONIBLES\n")
for col in df_ing.columns:
    print(f"\n==== {col} ====")
    if col == 'ingredient_name': 
        continue    
    # Tractament especial per a columnes amb llistes separades per |
    if col in ["base_flavors", "allowed_diets", "allergens"]:
        vals = set()
        df_ing[col].dropna().astype(str).str.split("|").apply(vals.update)
        vals = sorted(vals)
        print(f"(total {len(vals)} categories)")
        for v in vals:
            print(" -", v)
    else:
        uniques = sorted(df_ing[col].dropna().astype(str).unique())
        print(f"(total {len(uniques)} valors diferents)")
        # Imprimir tots (si n'hi ha pocs) o nomÃ©s una mostra si n'hi ha molts
        max_to_show = 50
        for v in uniques[:max_to_show]:
            print(" -", v)
        if len(uniques) > max_to_show:
            print(f" ... ({len(uniques) - max_to_show} mÃ©s)")
```
