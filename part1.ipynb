```python
class CasMenu: 
    def __init__(self, id_cas, problema, solucio, avaluacio):
        self.id_cas = id_cas
        self.problema = problema    # objecte DescripcioProblema
        self.solucio = solucio      # objecte SolucioMenu
        self.avaluacio = avaluacio  # objecte AvaluacioCas
        
        
class DescripcioProblema:
    def __init__(self, tipus_esdeveniment, formalitat, n_comensals, temporada, localitzacio, preu_per_persona, preferencies, restriccions_dietetiques=None): 
        self.tipus_esdeveniment = tipus_esdeveniment 
        self.formalitat = formalitat      # formal / informal
        self.n_comensals = n_comensals    # int    
        self.temporada = temporada        # estació de l'any
        self.localitzacio = localitzacio  # lloc o espai interior / exterior
        self.preu_per_persona = preu_per_persona  
        self.preferencies = preferencies  # llista de preferències culinàries
        self.restriccions_dietetiques = restriccions_dietetiques or []  # llista de restriccions dietètiques
        
class SolucioMenu: 
    def __init__(self, descripcio, diagnostics):
        self.descripcio = descripcio   
        self.diagnostics = diagnostics  
        # self.frase_represntativa = frase_representativa 
    
class AvaluacioCas:
    def __init__(self, derivacio, feedback, avaluacio, utilitat):
        self.derivacio = derivacio # origen del cas: cas real, derivat, etc
        self.feedback = feedback   # puntuacio estrelles i/o text amb comentaris
        self.avaluacio = avaluacio # exit o fracàs (ja veurem si boolea o puntuació)
        self.utilitat = utilitat  # valoració numèrica de la solució
```
